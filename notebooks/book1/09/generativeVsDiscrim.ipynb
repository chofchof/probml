# 9.4 Generative vs discriminative classifiers

## 9.4.1 Advantages of discriminative classifiers

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Figure 9.8: The class-conditional densities p(x|y=c) (left) may be more complex than the class posteriors p(y=c|x) (right). Adapted from Figure 1.27 of [Bis06].

```python
from scipy.stats import norm

def f1(x):
    return 0.6 * norm.pdf(x, loc=0.1, scale=0.09) + \
           0.4 * norm.pdf(x, loc=0.4, scale=0.09)

def f2(x):
    return norm.pdf(x, loc=0.7, scale=0.09)

def g1(x):
    return 1 / (1 + np.exp((27 * x - 15)))

def g2(x):
    return 1 / (1 + np.exp((-27 * x + 15)))

domain = np.linspace(0, 1, 100)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,4))
ax1.plot(domain, f1(domain), label='p(x|y=1)')
ax1.plot(domain, f2(domain), label='p(x|y=2)')
ax1.set(xlabel='x', ylabel='class conditional densities')
ax1.legend()

ax2.plot(domain, g1(domain), label='p(y=1|x)')
ax2.plot(domain, g2(domain), label='p(y=2|x)')
ax2.plot([0.556, 0.556], [0, 1.2], color='green')
ax2.set(xlabel='x', ylabel='class posteriors')
ax2.legend()

fig.tight_layout()
```
